## Step 4: Data Analysis
- Calculate daily average temperature
- Find max, min, mean for each metric
- Which hour of the day is the most humid on average?

```python
# TODO: Perform analysis


# Calculate daily average temperature
# convert the timestamp column 
df['timestamp'] = pd.to_datetime(df['timestamp'])
# extract the date first
df['date'] = df['timestamp'].dt.date
print(df['date'])
# Calculate daily average temperature
daily_avg_temp = df.groupby('date')['temperature_C'].mean()
print(daily_avg_temp)


#calculate max, min, mean of each matric
metrics = ['temperature_C', 'humidity_%', 'wind_speed_kmph', 'pressure_hPa', 'visibility_km']
df[metrics].agg(['max', 'min', 'mean', 'std'])


#which hour of the day is most humid on average?
df['hour'] = df['timestamp'].dt.hour
average_humid_hour = df.groupby('hour')['humidity_%'].mean()
most_humid_hour = average_humid_hour.idxmax()
print(f'The most Humid hour on average is {most_humid_hour}st hour')

```

### a) Reshape into matrix form
- Assume each row is a day
- Reshape temperature into a (10, 24) matrix
- Calculate daily min, max, and mean using axis-based operations

```python
# TODO: Reshape and aggregate

temp = df['temperature_C'].values
temp_matrix = temp.reshape((10, 24))
print(temp_matrix)
print("---------------------------------------------------")
# Write functions to find min, max, mean across rows
# Function to calculate daily minimum temperature
def daily_min(matrix):
    return np.min(matrix, axis=1)

# Function to calculate daily maximum temperature
def daily_max(matrix):
    return np.max(matrix, axis=1)

# Function to calculate daily average temperature
def daily_mean(matrix):
    return np.mean(matrix, axis=1)

print("Daily Minimum Temperatures:")
print(daily_min(temp_matrix))

print("\nDaily Maximum Temperatures:")
print(daily_max(temp_matrix))

print("\nDaily Average Temperatures:")
print(daily_mean(temp_matrix))
```

### b) Normalize the temperature matrix
- Subtract the mean and divide by std deviation
- Do it manually using NumPy functions

```python
# TODO: Normalize temp_matrix
def normalize(matrix):
  mean = np.mean(matrix)
  std = np.std(matrix)

  normalized_matrix = (matrix - mean) / std
  return normalized_matrix

normalized_temp = normalize(temp_matrix)
print(normalized_temp)
# return ...

# Apply it to temp_matrix
```

## Final Challenge: Write Your Own Function
Write a function `daily_summary(matrix)` that takes a NumPy matrix of shape (10, 24) and returns a summary dictionary for each day.

```python
# TODO: Write and test your function

#I write this 3 lines code to fill my missing values because in above sections where through mean, forward filling 
#and through bckward filling we have to fill the missing vales I made the copy of df and try it but
#I didn't change the original dataset
df['temperature_C'] = df['temperature_C'].fillna(df['temperature_C'].mean())
temp = df['temperature_C'].values
temp_matrix = temp.reshape((10, 24))

# Now i started to make a function and used that matrix

def daily_summary(temp_matrix):             
    summaries = []

    for day in temp_matrix:
        summary = {
            "min": float(np.nanmin(day)),
            "max": float(np.nanmax(day)),
            "mean": float(np.nanmean(day))
        }
        summaries.append(summary)    
    return summaries

summaries = daily_summary(temp_matrix)
for i, summary in enumerate(summaries, start=1):
    print(f"Day {i}: {summary}")
```
